```python
df1 = pd.read_csv("california_housing_train.csv")
```

```python
# for positive skewness log can be used to convert that skewness to symmetrical data
```

```python
df1["median_house_value_new"] = np.log(df1["median_house_value"])
```

```python
sns.distplot(df1["median_house_value_new"], fit = norm)
```

```python
df1["median_house_value"].skew()
```

```python
df1["median_house_value_new"].skew() #skew value should be near to zero
```

```python
#if the skewness is between -1 to -0.3 or 0.3 to 1 then data is considered as skewness. 
# in some rare case then the data is greater than 1 or lesser than -1 then the data is heavily skewed.  
```
